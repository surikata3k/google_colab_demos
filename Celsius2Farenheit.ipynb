```python
# Importem les llibreries
import tensorflow as tf
import numpy as np

# Especifiquem les dades d'entrada i de sortida
celsius = np.array([-40., -10, 0, 0.8, 15, 22, 38], dtype=float)
farenheit = np.array([-40, 14, 32, 46, 59, 72, 100], dtype=float)

# Cridem al framework Keras indicant que la nostra xarxa neuronal solament té
# una neurona (units = 1) i utilitzem un model seqüencial
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=1, input_shape=(1,))
])

# Ara compilem el nostre model utilitzant l'optimitzador Adam per a fer l'entrenament
# de l'aprenentatge automàtic. El número que se li passa com a paràmetre
# defineix els intervals de valors que agafarà. Un número més gran farà intervals
# més grans de pes i biaix
model.compile(
    optimizer=tf.keras.optimizers.Adam(0.1),
    loss='mean_squared_error'
)

# Anem a entrenar el model. Per a fer-ho utilitzem la funció fit, on ficarem les
# dades d'entrada, les de sortida i les voltes que volem que s'entreni (epochs)
print("Començant entrenament...")
historial = model.fit(celsius, farenheit, epochs=1000, verbose=False)
print("Model entrenat!!")
```

```python
import matplotlib.pyplot as plt
plt.xlabel("# Epochs")
plt.ylabel("Quantitat de pèrdua")
plt.plot(historial.history["loss"])
```

```python
print("Fem una predicció")
test = np.array([100], dtype=float)
resultat = model.predict(test)
print("El resultat és: " + str(resultat) + " Farenheit")

```
